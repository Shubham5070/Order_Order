```python
import pandas as pd

df_raw = pd.read_csv(
    "data/intent_data.csv",
    header=None,
    names=["raw"]
)

df_raw.head()
```

```python
# Drop rows that look like headers
df_raw = df_raw[~df_raw["raw"].str.lower().isin(["text,label", "text,intent"])]

df_raw.reset_index(drop=True, inplace=True)

df = df_raw["raw"].str.rsplit(",", n=1, expand=True)
df.columns = ["text", "intent"]

df.head()

```

```python
df["text"] = df["text"].str.strip().str.lower()
df["intent"] = df["intent"].str.strip().str.upper()

# Drop broken rows (just in case)
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

df.head()
```

```python
print("Total samples:", len(df))
print(df["intent"].value_counts())
```
